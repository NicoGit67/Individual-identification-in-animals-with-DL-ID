# Functions

```python
def create_translation_dict(string_list):
    unique_strings = sorted(set(string_list))
    translation_dict = {string: i for i, string in enumerate(unique_strings)}
    return translation_dict

def translate_to_integers(string_list, translation_dict):
    return [translation_dict[string] for string in string_list]

class_dict_train = create_translation_dict(y_train)
class_dict_val = create_translation_dict(y_val)
class_dict_test = create_translation_dict(y_test)
y_train = translate_to_integers(y_train, class_dict_train)
y_val = translate_to_integers(y_val, class_dict_val)
y_test = translate_to_integers(y_test, class_dict_test)
```

```python
def triplets_generator(dataset, batch_size): #, model_type = model_type):

    if dataset == 'train':
      gen_labels = y_train
    elif dataset == 'val':
      gen_labels = y_val

    def get_indices(lab):
        same_lab=np.where(gen_labels==lab)[0]
        id_anchor = np.random.choice(same_lab, replace=True) #TRY FALSE IF PROBLEM

        return id_anchor

    def create_batch(batch_size):
        batch_shape = (batch_size, sideH, sideW, n_channels)
        x_anchors = np.zeros(batch_shape)
        y_image = np.zeros((batch_size, n_classes))
        batch_lab=np.random.choice(gen_labels, batch_size, replace=True)

        for k, lab in enumerate(batch_lab):
            id_anchor = get_indices(lab)

            if dataset == 'train':
                x_anchors[k]= X_train[id_anchor,]

            if dataset == 'val':
                x_anchors[k]= X_val[id_anchor,]

            y_image[k, gen_labels[id_anchor]] = 1

        return x_anchors, y_image

    def data_generator(batch_size):
        while True:
            batch = create_batch(batch_size)

            yield batch

    tr_gen=data_generator(batch_size=batch_size)

    return tr_gen
```

# Model Construction

```python
if final_size == 'same':
  final_sideH, final_sideW = (sideH, sideW)
elif final_size != 'same':
  final_sideH, final_sideW = final_size

if model == 'InceptionV3':
  net = InceptionV3(
    include_top=True,
    weights=None,
    input_tensor=None,
    input_shape=(final_sideH, final_sideW, n_channels),
    pooling=None,
    classes=n_classes,
    classifier_activation='softmax')


  output = net.layers[-1].output
  net = Model(name = 'net', inputs = net.input, outputs = output)
  net._name ='net'
```

```python
n_epochs = 0

model_history  = pd.DataFrame()
print('no model found. initializing model')

input_anchor = tf.keras.layers.Input(name = 'input_anchor', shape=(sideH,sideW, n_channels))
x1 = input_anchor

if final_size != 'same':
  x1 = tf.image.resize(x1, final_size)

if augmentation == True:
  print('preprocessing images at the input of the model')

  x1 = RandomRotation(rotation_factor)(x1)
  x1 = RandomZoom(height_factor = h_factor, width_factor = w_factor)(x1)
  x1 = RandomBrightness(brightness_factor)(x1)
  x1 = RandomContrast(contrast_range)(x1)

output_classification = net(x1)

NN = tf.keras.models.Model(input_anchor, output_classification)
print('building classification model')

NN.compile(loss= 'categorical_crossentropy', metrics = ['accuracy'], optimizer='adamax')
print('compiling classification model')

#print(NN.summary())
```
